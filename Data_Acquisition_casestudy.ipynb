Q6: Identify any potential outliers in a numerical column of your choice. Explain
your approach.

```python
Q1 = numerical_column.quantile(0.25)
Q3 = numerical_column.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = numerical_column[(numerical_column < lower_bound) | (numerical_column > upper_bound)]

print(f"Potential outliers:\n{outliers}")
```

Q8: Group the data by a categorical column and compute the aggregate statistics
(e.g., mean, sum) for a numerical column.

```python

data = {
    'Job Title': ['Engineer', 'Designer', 'Engineer', 'Designer'],
    'Salary': ['$40.00 - $50.00 Per Hour (Employer est.)', '$60.00 - $70.00 Per Hour (Employer est.)', '$55.00 - $65.00 Per Hour (Employer est.)', '$70.00 - $80.00 Per Hour (Employer est.)']
}
df = pd.DataFrame(data)

def convert_salary_to_numeric(salary_str):
    salary_range = salary_str.split(' - ')
    if len(salary_range) == 2:
        min_salary = float(salary_range[0].replace('$', '').replace(' Per Hour (Employer est.)', ''))
        max_salary = float(salary_range[1].replace('$', '').replace(' Per Hour (Employer est.)', ''))
        return (min_salary + max_salary) / 2
    else:
        return float(salary_range[0].replace('$', '').replace(' Per Hour (Employer est.)', ''))

df['Salary'] = df['Salary'].apply(convert_salary_to_numeric)

aggregated_data = df.groupby('Job Title').agg({
    'Salary': ['mean', 'sum']
})

print(aggregated_data)


```
